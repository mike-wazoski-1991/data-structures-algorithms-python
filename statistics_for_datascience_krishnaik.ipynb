```python
#probably after third standard deviation, data will be an outlier
#Z-score
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
outliers = []
def detect_outliers(data):
  threshold = 3
  mean = np.mean(data)
  std = np.std(data)

  for i in data:
    z_score = (i-mean)/std
    if np.abs(z_score) > threshold:
      outliers.append(i)

  return outliers

k = detect_outliers(dataset)
# print(k)
plt.hist(dataset)
```

```python
#IQR
#1.Sort the data
#2. calculate Q1 and Q3
#3. IQR = Q3-Q1
#4. Lower Fence = Q1-1.5*IQR
#5. Upper Fence = Q3+1.5*IQR
sorted_dataset = sorted(dataset)
# sorted_dataset
q1,q3 = np.percentile(dataset,[25,75])#it will give us 25 and 75 percentile
# print(q1,q3)
iqr = q3-q1
# print(iqr)
lower_fence =q1-1.5*iqr
upper_fence = q3+1.5*iqr
print(lower_fence,upper_fence)
```

The Bernoulli distribution is a probability distribution for a random variable that can take only two possible outcomes, typically labeled as 1 (success) and 0 (failure). It is a discrete probability distribution and is defined by a single parameter,
𝑝
p, which represents the probability of success (1).

```python
import numpy as np
import matplotlib.pyplot as plt

# Probability of success (p)
p = 0.7

# Generate Bernoulli random variables (0 or 1)
data = np.random.binomial(n=1, p=p, size=1000)

# Plot the distribution
plt.hist(data, bins=2, edgecolor='black', rwidth=0.6)
plt.xticks([0, 1], labels=["Failure (0)", "Success (1)"])
plt.title(f"Bernoulli Distribution (p = {p})")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Probability of success (Head)
p = 0.7

# Outcomes and their probabilities
outcomes = ["Tail", "Head"]
probabilities = [1 - p, p]

# Plot the PMF
plt.bar(outcomes, probabilities, color=['red', 'green'], edgecolor='black', width=0.6)

# Add labels and title
plt.title("Bernoulli Distribution (PMF)")
plt.xlabel("Outcome")
plt.ylabel("Probability Mass Function (PMF)")
plt.ylim(0, 1)  # Limit the y-axis to [0, 1]
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate the bars with probabilities
for i, prob in enumerate(probabilities):
    plt.text(i, prob + 0.02, f"{prob:.2f}", ha='center', fontsize=10)

# Show the plot
plt.show()
```

A Binomial Distribution describes the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.



```python

import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the Binomial Distribution
n = 5  # Number of trials
p = 0.6  # Probability of success

# Possible values of X (number of successes)
x = range(n + 1)

# PMF of the Binomial Distribution
pmf = binom.pmf(x, n, p)

# Plot the PMF
plt.bar(x, pmf, color='skyblue', edgecolor='black', width=0.6)

# Add labels and title
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xlabel("Number of Heads")
plt.ylabel("Probability Mass Function (PMF)")
plt.xticks(x)  # Ensure the x-axis shows integer values
plt.ylim(0, max(pmf) + 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate the bars with probabilities
for i, prob in enumerate(pmf):
    plt.text(x[i], prob + 0.01, f"{prob:.2f}", ha='center', fontsize=10)

# Show the plot
plt.show()
```

The Pareto Distribution is a probability distribution that models wealth or resource distribution. It's often used in economics, social sciences, and business to describe situations where a small percentage of the population controls the majority of the wealth or resources. This is related to the 80/20 rule (e.g., 20% of individuals own 80% of the wealth).



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

# Parameters for the Pareto distribution
alpha = 2.5  # Shape parameter
xm = 10000   # Scale parameter (minimum value)

# Generate x values
x = np.linspace(xm, 5 * xm, 1000)  # From xm to 5*xm

# PDF and CDF
pdf = pareto.pdf(x - xm, alpha)  # Shift by xm
cdf = pareto.cdf(x - xm, alpha)

# Plot PDF
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)  # Subplot 1: PDF
plt.plot(x, pdf, color='blue', label='PDF')
plt.title(f"Pareto Distribution (α={alpha}, xₘ={xm})")
plt.xlabel('Income')
plt.ylabel('Probability Density')
plt.grid(alpha=0.5)
plt.legend()

# Plot CDF
plt.subplot(2, 1, 2)  # Subplot 2: CDF
plt.plot(x, cdf, color='green', label='CDF')
plt.xlabel('Income')
plt.ylabel('Cumulative Probability')
plt.grid(alpha=0.5)
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```

The Power Law Distribution is a mathematical relationship in which a relative change in one quantity results in a proportional relative change in another quantity, independent of the initial size of those quantities. It is commonly used to model distributions with heavy tails, such as wealth, file sizes, or the number of connections in a network.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Power Law distribution
alpha = 2.5  # Shape parameter
x_min = 1    # Minimum value (scale)

# Generate x values
x = np.linspace(x_min, 50, 1000)  # Range of values

# Calculate PDF
pdf = (alpha - 1) * (x_min ** (alpha - 1)) / (x ** alpha)

# Calculate CDF
cdf = 1 - (x_min / x) ** (alpha - 1)

# Plot PDF
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)  # Subplot 1: PDF
plt.plot(x, pdf, color='blue', label='PDF')
plt.title(f"Power Law Distribution (α={alpha}, x_min={x_min})")
plt.xlabel('Value (e.g., File Size)')
plt.ylabel('Probability Density')
plt.grid(alpha=0.5)
plt.legend()

# Plot CDF
plt.subplot(2, 1, 2)  # Subplot 2: CDF
plt.plot(x, cdf, color='green', label='CDF')
plt.xlabel('Value (e.g., File Size)')
plt.ylabel('Cumulative Probability')
plt.grid(alpha=0.5)
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```
